```python
import numpy as np
from pathlib import Path
```

```python
path2data = Path.home()/'group'/'project'/'results'/'beamforming'/\
'time_dependent'/'beamform_signals'/'220307_0to5cm_84to90_8192_gradb_corrected.npz'

name1 = '220328_87.0deg_500sec_range_17.5mm_5mm_range'
name2 = '220328_87.5deg_500sec_range_17.5mm_5mm_range'

file = np.load(path2data)

print(list(file.keys()))

pitch_angles = file['pitch']
radii = file['radius']
print(pitch_angles.shape)
print(file['signals'].shape)

pitch_range_1 = [86.75, 87.25]
pitch_range_2 = [87.25, 87.75]
radius_range = [0.015, 0.02]

unique_pitch = np.unique(pitch_angles)
print(np.unique(radii))

select_ind_1 = np.argwhere(np.logical_and(
                                np.logical_and(
                                    pitch_angles <= pitch_range_1[1],
                                    pitch_angles >= pitch_range_1[0]
                                ),
                                np.logical_and(
                                    radii <= radius_range[1],
                                    radii >= radius_range[0]
                                )
                            )).squeeze()

select_ind_2 = np.argwhere(np.logical_and(
                                np.logical_and(
                                    pitch_angles <= pitch_range_2[1],
                                    pitch_angles >= pitch_range_2[0]
                                ),
                                np.logical_and(
                                    radii <= radius_range[1],
                                    radii >= radius_range[0]
                                )
                            )).squeeze()

select_signal_1 = file['signals'][select_ind_1, :]
select_signal_2 = file['signals'][select_ind_2, :]

print(select_signal_1.shape, select_signal_2.shape)

select_nonzero_inds_1 = np.argwhere(abs(select_signal_1).sum(-1) > 0).squeeze()
select_nonzero_inds_2 = np.argwhere(abs(select_signal_2).sum(-1) > 0).squeeze()

select_signal_1 = select_signal_1[select_nonzero_inds_1]
select_signal_2 = select_signal_2[select_nonzero_inds_2]

print(select_signal_1.shape, select_signal_2.shape)


save_path = Path.home()/'group'/'project'/'results'/'beamforming'/\
'time_dependent'/'beamform_signals'

np.save(save_path/name1, select_signal_1)
np.save(save_path/name2, select_signal_2)

```
