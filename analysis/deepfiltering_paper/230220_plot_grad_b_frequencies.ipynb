```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

gradb_data = Path.home()/'group'/'project'/'results'/'mayfly'/'211129_grad_b_frequency_grid_radius_angle.npz'
gradb_data = np.load(gradb_data)
```

```python
print(list(gradb_data.keys()))
```

```python
print(gradb_data['freq'].shape)
```

```python
sns.set_theme(context='talk', style='ticks', font_scale=1.2)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
#cmap = sns.color_palette('coolwarm', as_cmap=True)
#cmap = sns.diverging_palette(120, 5, s=90, l=50, center='light', as_cmap=True)
cmap = sns.diverging_palette(5, 255, s=100, l=50, center='light', as_cmap=True)
img = ax.imshow(
    gradb_data['freq'][:, 0:50]*2*np.pi,
    aspect='auto',
    cmap=cmap,
    interpolation='nearest',
    extent=(0,50,90,84)
)

cbar = fig.colorbar(img, label='Drift Frequency (rad/s)')
ax.set_xticks(np.linspace(0,50,6))
ax.set_xticklabels(np.arange(0,60,10))
ax.set_xlabel('Radial Position (mm)')
ax.set_ylabel('Pitch Angle (deg)')

ax.set_title('Guiding Center Drift Frequency vs.'+'\n'+'Radial Position and Pitch Angle')
plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'plots'/'230220_gradb_drift_frequencies')
```
