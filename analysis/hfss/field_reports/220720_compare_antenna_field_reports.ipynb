```python
import csv
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
home = Path.home()
dipole_fields = home/'group'/'project'/'results'/'hfss'/'220302_cres_antenna_fields'/'220302_cres_dipole_rE_phi_sweep.csv'
synca_fields = home/'group'/'project'/'results'/'hfss'/'220414_cres_antenna_asbuilt_fields'/'220414_cres_antenna_asbuilt_fields_phi.csv'


def ParseCSV(file):
    
    df = pd.read_csv(file)
    #with open(file, 'r') as infile:
    #for i, line in enumerate(csv.reader(infile)):
    #       print(line)
    
    return df
```

```python
data_list = []


for file in [dipole_fields, synca_fields]:
    
    frame = ParseCSV(file)
    data_list.append(frame)
```

```python
sns.set_theme(context='talk', style='whitegrid', font_scale=1.2, palette='deep')
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1,1,1, projection='polar')

plot_keys0 = ['mag(rETotal) [mV]']
plot_keys1 = ['mag(rETotal) [V]']
plot_keys = [plot_keys0, plot_keys1]

labels = ['Crossed Dipole', 'SYNCA']
for n, index in enumerate([0, 1]):

    theta_sweep = data_list[index]

    thetas = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = plot_keys[n][0]).mean().index.array)
    
    for m, key in enumerate(plot_keys[n]):
        
        re_comp = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = key).mean().array)
        if m == 0:
            norm = np.max(re_comp)
            
        trace = 10 * np.log10(re_comp / norm)
        print(np.mean(trace), np.std(trace), np.max(trace), np.min(trace), np.max(trace)-np.min(trace))

        ax.plot(np.radians(np.arange(0, 362, 2)), trace, label=labels[n], lw=3,)

ax.plot(np.radians(np.arange(0, 360, 1)), np.zeros(360), label='Sim. Electron', color='k', linestyle=':', lw=4)
#ax.set_thetamin(-90)
#ax.set_thetamax(90)
#ax.set_theta_zero_location('N')
#ax.set_theta_direction(-1)

ax.set_rmin(-6)
ax.set_rmax(1)

ax.legend(loc=(0.26,0.2))
ax.set_title('Normalized Electric Field Magnitude ')

#xticks = ax.set_xticks(np.radians(np.arange(90, -120, -30)))
#xticklabels = ax.set_xticklabels(labels=[fr'{n}$^\circ$' for n in np.arange(0, 210, 30)])
#ax.set_xlabel(r'Zenith Angle (deg)', loc='center', y=100 )
ax.text(np.radians(135), -2 ,r'Elevation = 90$^\circ$')
ax.text(np.radians(12), 0, 'Magnitude (dB)', rotation=-65)

plt.tight_layout()
name = '220728_synca_hfss_polar_comparison_db.png'
save = home/'group'/'project'/'plots'/'hfss'/'synca'
save.mkdir(parents=True, exist_ok=True)

#plt.savefig(save/name, dpi=128)
#####


'''
sns.set_theme(context='talk')
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1,1,1, projection='polar')
for n, index in enumerate([0]):
    
    theta_sweep = data_list[index]

    phis = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = plot_keys[n]).mean().index.array)
    retotal = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = plot_keys[n]).mean().array)
    
    #phis = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = plot_keys[n]).index.array)
    #retotal = np.array(theta_sweep.pivot(index='Theta [deg]', columns='Phi [deg]', values = plot_keys[n]))

    #print(retotal)
    if plot_keys[n].find('pV') > 0:
        ax.plot(np.radians(phis), retotal / 1e3, label=labels[n])
    else:
        ax.plot(np.radians(phis), retotal, label=labels[n])

ax.set_rmin(0)
ax.set_rmax(10)
ax.legend()
ax.set_title('Total Electric Field, Zenith Angle Average')
ax.text(np.radians(-20), 12.,'Azimuthal Angle (deg)', rotation=270)
ax.text(np.radians(65), 1.5, 'Electric Field (mV/m)', rotation=25)
#ax.set_xlabel('Azimuthal Angle (deg)')

plt.tight_layout()
name = '220502_pcb_cross_dipole_field_components_hfss_azimuth.png'
save = home/'group'/'project'/'plots'/'hfss'/'pcb_cross_dipole'
save.mkdir(parents=True, exist_ok=True)
plt.savefig(save/name)

'''
```
