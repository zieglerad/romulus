```python
import moth
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import scipy
import scipy.signal

PATH = '/storage/home/adz6/group/project'
VNADATAPATH = os.path.join(PATH, 'vna_data')
DAQDATAPATH = os.path.join(PATH, 'daq_data')
MOTHPATH = os.path.join(PATH, 'moth')
MOTHDATA = os.path.join(MOTHPATH, 'data/datasets')
PLOTPATH = os.path.join(PATH, 'plots/moth')
MOTHRESULTS = os.path.join(PATH, 'results/moth')

def LoadData(path2data, nskip_start=10240, remove_mean=True, convert_to_volts=True):
    
    nantenna = len(os.listdir(path2data))
    antenna_angles = []
    for iantenna, antenna in enumerate(os.listdir(path2data)):
        
        antenna_angles.append(int(antenna.split('.csv')[0]))
        
        _array = LoadArray(os.path.join(path2data, antenna))
        if remove_mean:
            _array = _array - np.mean(_array, axis = -1).reshape((_array.shape[0], 1)).repeat(_array.shape[-1], axis=-1)

        if iantenna == 0:
            antenna_data = np.zeros((nantenna, _array.shape[0]-1, _array.shape[-1]))
            lo_data = np.zeros((nantenna, _array.shape[-1]))
        
        antenna_data[iantenna, :, :] = _array[1:, :]
        lo_data[iantenna, :] = _array[0, :]
                
    antenna_sort_inds = np.argsort(antenna_angles)
    
    antenna_angles = -1 * np.sort(antenna_angles)
    
    antenna_data = scipy.signal.hilbert(antenna_data[antenna_sort_inds, :, :], axis=-1)
    lo_data = scipy.signal.hilbert(lo_data[antenna_sort_inds, :], axis=-1)
    
    if convert_to_volts:
        antenna_data *= 0.5 / 2**16
        lo_data *= 5 / 2**16
    
    return {'angle': antenna_angles, 'antenna': antenna_data, 'lo': lo_data}

    
def LoadArray(path2array, nskip_start=10240):
    
    signal_list = []
    with open(path2array, 'r') as infile:
        reader = csv.reader(infile)
        for i, item in enumerate(reader):
            #print(i, np.int32(np.array(item)))
            _array = np.int32(np.array(item))
            if _array.sum() > 0:
                signal_list.append(_array)
    return np.array(signal_list)[:, nskip_start:]



def CorrectLOPhase(array_data):
    
    lo_fft = np.fft.fft(array_data['lo'])
    
    phase_at_max = np.diag(np.angle(lo_fft[:, np.argmax(abs(lo_fft), axis=-1)]))
    
    phase_diff = phase_at_max - phase_at_max[0]
    
    correction = np.exp(-1j * phase_diff).reshape((*phase_diff.shape, 1)).repeat(lo_fft.shape[-1], axis=-1)
    
    array_data['lo'] *= correction
    correction = correction.reshape((correction.shape[0], 1, correction.shape[-1])).repeat(array_data['antenna'].shape[1], axis=1)
    array_data['antenna'] *= correction
    

def GetPhase(array):
    data_fft = np.fft.fft(array, axis=-1)
    #print(data_fft.shape)
    max_inds = np.argmax(abs(data_fft[:, 1, :]), axis=-1)
    phase_array = np.diag(np.angle(data_fft[:, 1, max_inds]))
    
    return phase_array



def ComputeMagCorrection(array_data):
    
    antenna_fft = np.fft.fft(array_data['antenna'], axis=-1)
    
    for channel in range(array_data['antenna'].shape[1]):
        
        angle_magnitudes = abs(antenna_fft[:, channel, :]).sum(axis=-1)
        
        angle_magnitudes /= np.max(angle_magnitudes)
        
        if channel == 0:
            mag_correction = np.zeros((array_data['antenna'].shape[1], array_data['antenna'].shape[0]))
            
        mag_correction[channel, :] = 1 / angle_magnitudes
        
    return mag_correction

def ComputePhaseCorrection(array_data):
    
    antenna_fft = np.fft.fft(array_data['antenna'], axis=-1)
    
    for channel in range(array_data['antenna'].shape[1]):
        
        phases_at_max = np.angle(antenna_fft[:, channel, np.argmax(abs(antenna_fft[0, channel, :]))])
        
        #angle_magnitudes /= np.max(angle_magnitudes)
        
        if channel == 0:
            phase_correction = np.zeros((array_data['antenna'].shape[1], array_data['antenna'].shape[0]))
            
        phase_correction[channel, :] = phases_at_max - phases_at_max[0] - np.radians(np.arange(0, 360, 360 // antenna_fft.shape[0]))# relative phases + anti-spiral
        
    return phase_correction

def ApplyMagCorrection(array_data, mag_correction):
    
    for channel in range(mag_correction.shape[0]):
        
        array_data['antenna'][:, channel, :] *= mag_correction[channel, :].reshape(mag_correction.shape[-1], 1).repeat(array_data['antenna'].shape[-1], axis=-1)
        
def ApplyPhaseCorrection(array_data, phase_correction):
    
    for channel in range(phase_correction.shape[0]):
        array_data['antenna'][:, channel, :] *= np.exp(-1j * phase_correction[channel, :]).reshape(phase_correction.shape[-1], 1).repeat(array_data['antenna'].shape[-1], axis=-1)
```
