```python
## imports ##
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle as pkl
import MFTime
import MFParse
import os
import scipy.signal
import seaborn as sns
####

## global ##
dataPath='/Users/ziegler/repos/mayfly/output/timeSeries1252021.pkl'
simPath='/Users/ziegler/iCloud/egg_files/eggsTemplates1252021/'
templatePitchAngles=np.linspace(85,90,51)
templatePos=np.linspace(0,5e-2,21)
radius=0.00
#nPeaks=5
nTest=100
noiseVar=2.3e-14
colors=['r','b','g','c','m','k']
frequencyConversion=200e6/8192
array_rad=0.1
Vrange=(80-280*array_rad+300*array_rad**2)*1e-9
nSlice=2
sliceSize=8192
samplePeriod=5e-9
time=np.arange(0,sliceSize)
####

## definitions ##

def getEggFilePaths(dirPath,listOfPaths):
    try:
        os.path.isdir(dirPath)
    except:
        return False

    for fPath in os.listdir(dirPath):
        if fPath.split('.')[-1]=='egg':
            listOfPaths.append(os.path.join(dirPath,fPath))

    return True

def getEggFileParams(eggPath,params):
    try:
        os.path.isfile(eggPath)
    except:
        return False

    try:
        pitchAngle=float(eggPath.split('/')[-1].split('Angle')[-1].split('_')[0])
        radius=float(eggPath.split('/')[-1].split('Pos')[-1].split('.egg')[0])
    except:
        return False

    params.append(pitchAngle)
    params.append(radius)

    return True

def getSimpleTemplate(frequencies,amplitudes,relPhase,nSample,dT):

    simpleTemplate=np.zeros(nSample,dtype=np.complex128)
    time=np.arange(0,nSample,1)*dT
    maxAmplitude=np.argmax(amplitudes)
    for i in range(len(frequencies)):
        if i == maxAmplitude:
            simpleTemplate+=amplitudes[i]*np.exp(+1j*2*np.pi*(frequencies[i]*time))
        else:
            simpleTemplate+=amplitudes[i]*np.exp(+1j*2*np.pi*(frequencies[i]*time+relPhase))

    return simpleTemplate

def addNoise(input,var):

    noise=np.random.multivariate_normal([0,0],np.eye(2)*var/2,len(input))
    noise=noise[:,0]+1j*noise[:,1]

    return input+noise

def getSimpleFrequencyEstimate(frequencySeries,threshold=0.25):

    maxValue=np.max(abs(frequencySeries))
    maxInd=np.argmax(abs(frequencySeries))
    #print(maxInd)

    rawPeakInds=np.where(abs(frequencySeries)>threshold*maxValue)[0]
    
    bigPeakInd=np.flip(np.argsort(abs(frequencySeries)))[0:10]
    bigPeakAmp=abs(frequencySeries[bigPeakInd])
    
    groupedPeakIndInd=np.argsort(bigPeakInd)
    
    groupedPeakInd=bigPeakInd[groupedPeakIndInd]
    groupedPeakAmp=bigPeakAmp[groupedPeakIndInd]

    return groupedPeakInd,groupedPeakAmp

def GenerateSimpleAndOmniTemplates(simpleTemplateList, OmniTemplateList, eggPath, relativePhase=np.linspace(0,1,10),
                            Vrange=5.5e-8, sliceSize=8192, nSlice=2, samplePeriod=5e-9):
    
            eggDataTime=MFParse.parse_egg(eggPath,Vrange=Vrange)
            eggSliceTime=MFParse.slice_egg(eggDataTime,nSlice,sliceSize)
            
            if MFParse.check_egg_slice(eggSliceTime): # check if the simulation is broken
                badEgg.append(eggPath)
            else:
                simulationParams=[]
                getEggFileParams(eggPath,simulationParams)
                
                eggSliceFFT,FFTFrequencies=MFTime.katydid_fft(eggSliceTime)
                trueSignalTime=eggSliceTime[0,:]
                trueSignalFreq=eggSliceFFT[0,:]/sliceSize
                
                freqInd,freqAmp=getSimpleFrequencyEstimate(trueSignalFreq)
                freq=FFTFrequencies[freqInd]
                
                temp = {'type': 'omni', 'pitch-angle': simulationParams[0], 'radius': simulationParams[1],
                            'time-series': trueSignalTime, 'freq-series': trueSignalFreq}
                
                OmniTemplateList = [temp]
                
                for i, phase in enumerate(relativePhase):
                    simpleTemplate=getSimpleTemplate(freq,freqAmp,phase,sliceSize,samplePeriod)
                    simpleTemplateFFT=np.fft.fftshift(np.fft.fft(simpleTemplate))
                    
                    temp = {'type': 'simple', 'pitch-angle': simulationParams[0], 'radius': simulationParams[1],
                            'time-series': simpleTemplate, 'freq-series': simpleTemplateFFT}
                    
                    simpleTemplateList.append(temp)
                    
        return True
    
def GenerateNoisySignal(signalEntryDictList, templateDictList, var):
    
    
    signalNoise = addNoise(templateDict['time-series'], var)
    temp = {'type': templateDict['type'], 'pitch-angle': templateDict['pitch-angle'],
            'radius': templateDict['radius'], 'time-series': signalNoise,
            'freq-series': np.fft.fftshift(np.fft.fft(signalNoise))}
    
    signalEntryDict = temp
    
    return True
    
def CalculateMFInnerProduct(signalEntryDict, templateDictList, mfInnerProductDictList, NSample=8192):
    
    for i, entry in enumerate(templateDictList):
        temp = {'signal-type': signalEntryDict['type'], 'signal-pitch-angle': signalEntryDict['pitch-angle'],
                'signal-radius': signalEntryDict['radius'], 'template-type': templateDictList[i]['type'],
                'template-pitch-angle': templateDictList[i]['pitch-angle'],
                'template-radius': templateDictList[i]['radius']}
        
        innerProduct = abs(np.sum(signalEntryDict['freq-series'] * templateDictList[i]['freq-series']))/NSample
        
        temp.update('inner-product': innerProduct)
        
        mfInnerProductDictList.append(temp)
        
        
def CalculateInnerProducts(simulationPath):
    
    eggPaths = []
    Inner
    if not getEggFilePaths(simulationPath, eggPaths):
        print('Failed to find egg files.')
        return False
    else:
        
        for i, eggPath in enumerate(eggPaths):
            
            omniTemplate = []
            simpleTemplates = []
            GenerateSimpleAndOmniTemplates(simpleTemplates, omniTemplate, eggPath)
            
            noisySignal = []
            GenerateNoisySignal(noisySignal, omniTemplate)
            
            
            
            
        
        
####
```
