# cosmology

```python
sns.set_theme(style='ticks', context='paper', font_scale=1.75)
clist = sns.color_palette('bright', n_colors=10)
plt.rcParams['text.usetex']=False
```

```python
def msum_normal(mleast, m21, m31, ):
    return mleast + np.sqrt(m21**1 + mleast**2) + np.sqrt(m31**1 + mleast**2)

def msum_invert(mleast, m31, m32, ):
    return mleast + np.sqrt(-m32**1 + mleast**2) + np.sqrt(-m31**1 + mleast**2)

# nu-fit 5.2 w/o SK
m21_mean_normal = 7.41e-5 #ev^2
m31_mean_normal = 2.51e-3 #ev^2

m32_mean_invert = -2.498e-3 
m21_mean_invert = 7.41e-5
m31_mean_invert = m32_mean_invert + m21_mean_invert

mleast = np.linspace(0, 1, 1001)


fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1,1,1)

#ax.plot(mleast, mleast)
ax.plot(
    mleast,
    msum_normal(mleast, m21_mean_normal, m31_mean_normal),
    color= clist[0],
    lw=2,
    label='Normal'
)
ax.plot(
    mleast,
    msum_invert(mleast, m31_mean_invert, m32_mean_invert),
    color=clist[1],
    lw=2,
    label='Inverted'
)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(3e-2, 2e0)

ax.set_xlabel(r'$m_\mathrm{least}$ (eV)')
ax.set_ylabel(r'$\sum_i{m_i}$ (eV)', rotation=0)
#ax.set_title(r'Cosmological Neutrino Mass Observable'+'\n'+' as a Function of $m_\mathrm{least}$')
ax.legend(loc=2)
plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'plots'/'230301_cosmology_nu_mass_observable', dpi=144)

```

# best fit mass e-values

```python
def mass3_n(m):
    
    m31_2_bf = 2.51e-3
    
    return np.sqrt(m31_2_bf + m **2)

def mass2_n(m):
    
    m21_2_bf = 7.41e-5
    
    return np.sqrt(m21_2_bf + m **2)

def mass1_i(m):
    
    m32_2_bf = -2.498e-3 
    m21_2_bf= 7.41e-5
    m31_2_bf = m32_2_bf + m21_2_bf
    
    return np.sqrt(m **2 - m31_2_bf)

def mass2_i(m):
    
    m32_2_bf = -2.498e-3 
    
    return np.sqrt(m **2 - m32_2_bf)

fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,1,1)
#ax2 = fig.add_subplot(1,2,2)

m = np.linspace(0,1.2, 1001)

# normal
ax1.plot(m, m, lw=2, color=clist[0], label=r'$m_1$')
ax1.plot(m, mass2_n(m), lw=2, color=clist[1], label=r'$m_2$')
ax1.plot(m, mass3_n(m), lw=2, color=clist[2], label=r'$m_3$')


ax1.set_xscale('log')
ax1.set_yscale('log')

ax1.set_ylim(0.001, 0.12)
ax1.set_xlim(0.001, 0.12)

ax1.set_xlabel(r'$m_\mathrm{least}$ (eV)')
ax1.set_ylabel('Mass (eV)')
ax1.legend(loc=4)
ax1.set_title('Normal Mass Ordering')

plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'plots'/'230302_mass_estate_vals_normal', dpi=144)


fig = plt.figure(figsize=(8,5))
ax2 = fig.add_subplot(1,1,1)


# invert
ax2.plot(m, mass1_i(m), lw=2, color=clist[0], label=r'$m_1$')
ax2.plot(m, mass2_i(m), lw=2, color=clist[1], label=r'$m_2$')
ax2.plot(m, m, lw=2, color=clist[2], label=r'$m_3$')


ax2.set_xscale('log')
ax2.set_yscale('log')

ax2.set_ylim(0.001, 0.12)
ax2.set_xlim(0.001, 0.12)

ax2.set_xlabel(r'$m_\mathrm{least}$ (eV)')
ax2.set_ylabel('Mass (eV)')
ax2.legend(loc=4)
ax2.set_title('Inverted Mass Ordering')

plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'plots'/'230302_mass_estate_vals_inverted', dpi=144)

```

# double beta decay mass observable in terms of beta-decay mass

```python
from matplotlib.patches import Patch

# normy

def mbb_max_n(m):
    
    return 0.681 * np.sqrt(m**2 - 7.74e-5) + 0.297 * np.sqrt(m**2 - 3.3e-6) + 0.022 * np.sqrt(m**2 + 2.43e-3)
m = np.linspace(7.74001e-5,1.2, 10001)

def mbb_min_n(m):
    
    return 0.681 * np.sqrt(m**2 - 7.74e-5) - 0.297 * np.sqrt(m**2 - 3.3e-6) - 0.022 * np.sqrt(m**2 + 2.43e-3)

# inverty

def mbb_max_i(m):
    
    return 0.681 * np.sqrt(m**2 - 4.61e-5)  + 0.296 * np.sqrt(m**2 + 2.8e-5) + 0.022 * np.sqrt(m**2 - 2.47e-3)


def mbb_min_i(m):
    
    return 0.681 * np.sqrt(m**2 - 4.61e-5)  - 0.296 * np.sqrt(m**2 + 2.8e-5) - 0.022 * np.sqrt(m**2 - 2.47e-3)

m_i = np.linspace(2.47e-3,1.2, 10001)
m_n = np.linspace(7.74001e-5,1.2, 10001)



fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)

ax.plot(m_n, mbb_max_n(m_n), color=clist[0], lw=2)
ax.plot(m_n, mbb_min_n(m_n), color=clist[0], lw=2)
ax.fill_between(m_n, mbb_min_n(m_n), mbb_max_n(m_n), color=clist[0], alpha=0.33, ls='-',)

ax.plot(m_i, mbb_max_i(m_i), color=clist[1], lw=2)
ax.plot(m_i, mbb_min_i(m_i), color=clist[1], lw=2)
ax.fill_between(m_i, mbb_min_i(m_i), mbb_max_i(m_i), color=clist[1], alpha=0.33, ls='-',)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-3, 1.2)

patch_normal = Patch(ec=clist[0], fc='#A3B9FE', alpha=1, ls='-', lw=2)
patch_inverted = Patch(ec=clist[1], fc='#FDCBA3', alpha=1, ls='-', lw=2)

ax.legend([patch_normal, patch_inverted],['Normal', 'Inverted'], loc=2)

ax.set_xlabel(r'$m_\beta$ (eV)')
ax.set_ylabel(r'$|m_{\beta\beta}|$ (eV)')
plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'plots'/'230301_mbb_vs_mb')
```
