```python
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

def trajectory(t,a0,a1,a2,w0,w1,w2):
    
    x = a0*np.cos(w0*t) + a2*np.cos(w2*t)
    y = a0*np.sin(w0*t) + a2*np.sin(w2*t)
    z = a1*np.cos(w1*t)
    
    return x,y,z

a0 = 3e-3
a1 = 1e-2
a2 = 1e-1

w0 = 1.25e2
w1 = 1
w2 = 1e-1

npts = np.int32(1e7)
t = np.linspace(0, 2*np.pi/w2, npts)

x,y,z = trajectory(t,a0,a1,a2,w0,w1,w2)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot(
    x,y,z,
    lw=0.75,
    color='k'
)

ax.set_zlim(-0.025,0.025)

ax.plot(
    np.linspace(-0.005,0.05, 100),
    np.zeros(100),
    np.zeros(100)
)

ax.plot(
    np.zeros(100),
    np.linspace(-0.005,0.05, 100),
    np.zeros(100)
)

ax.plot(
    np.zeros(100),
    np.zeros(100),
    np.linspace(-0.001,0.01, 100),
)

ax.set_axis_off()

plt.tight_layout(pad=1.05)

plt.savefig(Path.home()/'group'/'project'/'230511_electron_trajectory', dpi=2048)
```
